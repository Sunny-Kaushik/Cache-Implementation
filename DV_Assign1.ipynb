```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

import seaborn as sns


df = pd.read_csv("rural_data.csv")
month_name_mapping = {
    "January": "Jan",
    "February": "Feb",
    "March": "Mar",
    "April": "Apr",
    "May": "May",
    "June": "Jun",
    "July": "Jul",
    "August": "Aug",
    "September": "Sep",
    "October": "Oct",
    "November": "Nov",
    "December": "Dec"
}
df['Name'] = df['Name'].map(month_name_mapping)
# Combine 'year' and 'month' columns into a single 'date' column
df['date'] = df['Year'].astype(str) + '-' + df['Name']  # Combine year and month
df['date'] = pd.to_datetime(df['date'], format='%Y-%b')  # Convert to datetime with abbreviated month name format

# Define the state you want to plot
state_to_plot = 'Delhi'  # Replace with the desired state name

# Pivot the DataFrame to have states in rows
df_pivot = df.melt(id_vars=['date', 'Year', 'Name'], var_name='state', value_name='CPI')

# Filter data for the specific state
state_data = df_pivot[df_pivot['state'] == state_to_plot]

# Create a line plot for the specific state
plt.figure(figsize=(20, 7))
sns.lineplot(data=state_data, x='date', y='CPI')
plt.title(f"CPI Trend for {state_to_plot}")
plt.xlabel("Date")
plt.ylabel("CPI")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

```python


# Load the CSV file into a DataFrame
df = pd.read_csv("rural_data.csv")

# Mapping for month names
month_name_mapping = {
    "January": "Jan",
    "February": "Feb",
    "March": "Mar",
    "April": "Apr",
    "May": "May",
    "June": "Jun",
    "July": "Jul",
    "August": "Aug",
    "September": "Sep",
    "October": "Oct",
    "November": "Nov",
    "December": "Dec"
}
df['Name'] = df['Name'].map(month_name_mapping)

# Combine 'Year' and 'Name' columns into a single 'date' column
df['date'] = df['Year'].astype(str) + '-' + df['Name']  # Combine year and month
df['date'] = pd.to_datetime(df['date'], format='%Y-%b')  # Convert to datetime with abbreviated month name format

# Define the state you want to plot
state_to_plot = 'Delhi'  # Replace with the desired state name

# Pivot the DataFrame to have states in rows
df_pivot = df.melt(id_vars=['date', 'Year', 'Name'], var_name='state', value_name='CPI')

# Filter data for the specific state
state_data = df_pivot[df_pivot['state'] == state_to_plot]

# Create a line plot for the specific state
plt.figure(figsize=(35, 6))
sns.lineplot(data=state_data, x='date', y='CPI')
plt.title(f"CPI Trend for {state_to_plot}")
plt.xlabel("Date (Month)")
plt.ylabel("CPI")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability

# Format the x-axis labels to display full month names and years
date_labels = state_data['date'].dt.strftime('%B %Y')
plt.xticks(ticks=state_data['date'], labels=date_labels)

plt.show()
```
